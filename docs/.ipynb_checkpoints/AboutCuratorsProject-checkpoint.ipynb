
# Кураторская методика 
### Общая информация о проекте

[«Кураторская методика»](https://direktoria.org/academy/kompleksnye-programmy-vnedreniya-izmeneniy-v-shkolakh.php) - это проект, развиваемый Издательской фирмой «Сентябрь» на своей образовательной платформе «Академия Директории», являющейся частью информационной системы Директория.

Обучение по этой программе организуется для руководителей школ, администраторов и методистов школ, выполняющих функции кураторов, а также педагогов, которые выбраны руководством школы для участия в проекте. Суть методики заключается в том, что в школе, которая принимает участие в проекте, создаются "тройки" (одна или больше), включающие в себя куратора и двух педагогов. Куратор выполняет роль модератора и помощника, а педагоги, имея равные права, по очереди посещают уроки друг друга. Для каждого такого урока пара педагогов совместно с куратором определяют один из доступных в рамках проекта протоколов наблюдения. Такой протокол позволяет зафиксировать отдельные параметры урока, соответствующие тому педагогическому навыку, который на уроке должен отработать проводящий урок учитель. 
После проведения урока и заполнения протокола учителя совместно с куратором, основываясь на данных, зафиксированных в протоколе, обсуждают то, насклько удачно или неудачно с точки зрения отрабатываемого навыка был проведен урок, фиксируют проблемы, отмечают удачи и планируют свои дальнейшие действия.

Главными целями проекта являются:  

*   Повышение профессионального уровня педагогов;
*   Повышение уровня внимания педагогов к отстающим ученикам и, как следствие, повышение среднего уровня качества обучения;
*   Формирование горизонтальных профессиональных связей внутри организации - повышение уровня социального капитала организации.


Обучение и методическая поддержка процесса обеспечиваются за счет:
*   самостоятельной работы участников проекта с учебными материалами;
*   участия в вебинарах и онлайн-тренировках;
*   коммуникаций с администраторами Директории, курирующими процесс обучения.


В 2021 году в рамках этого проекта было проведено обучение школ с низкими образовательными результатами Новосибирской области. Руководство регионального министерства образования, убедившись в эффективности методики, приняло решение продлить взаимодействие с «Сентябрем» и в 2022 году, и начиная с 2022 года проект получил новое развитие. Начиная с января 2022 года команды из 67 школ Новосибирской области включились в проект и в течение всего 2022 года учителя и кураторы этих школ провели в общей сложности более 2500 уроков.


Весной началась работа над созданием административной панели проекта "Кураторская методика", первоначальная цель создания которой - автоматизация работы по сбору результатов наблюдений на уроках и формирования инструментов для анализа накапливаемых данных и предоставления участникам обучения аналитики по результатам апробации кураторской методики.

Благодаря созданию этого сайта ([http://kurators.direktoria.org](http://kurators.direktoria.org)) в течение 2022 года был накоплен объем данных, обработка и анализ которого и планируется сделать основной целью выполнения этой дипломной работы.
